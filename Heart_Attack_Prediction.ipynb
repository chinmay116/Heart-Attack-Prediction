```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# read an excel file and convert into a dataframe object 
df = pd.DataFrame(pd.read_excel("data.xlsx")) 
```

```python
#split features and target
# X -> feaures
# Y -> target

X = df.drop(columns='target', axis=1)

Y = df['target']
```

```python
X.shape
```

```python
Y
```

```python
#train & test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=5)
```

```python
X.shape
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
#train the model
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#model accuracy score (Training)
X_train_predict = model.predict(X_train)
data_accuracy = accuracy_score(X_train_predict, Y_train)
```

```python
X_train_predict      #displaying pridicted target values based on the X_train
```

```python
data_accuracy
```

```python
#model accuracy score (Testing)
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict, Y_test)
```

```python
X_test_predict        #displaying pridicted target values based on the X_test
```

```python
test_data_accuracy
```
